```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
df = pd.read_excel('draws.xlsx', 'total')
```

```python
df.describe()
```

```python
df_old = df.drop([12])
```

```python
sns.lineplot(data=[df_old['goals']/462, df_old['draws']/462])
```

```python
df_old['avg_goals'] = df_old['goals'] / 462
df_old['avg_draws'] = df_old['draws'] / 462
```

```python
sns.regplot(data=df_old, x="season_start", y="draws")
```

```python
sns.regplot(data=df_old, x="season_start", y="avg_goals")
```

```python
fig, ax = plt.subplots()

plt.axvline(1981, 0, 4, color='black', linestyle='--', linewidth=1)

sns.regplot(data=df_old, x="season_start", y="avg_goals", marker='o', ax=ax)

ax.set_ylim(0, 4) 
plt.show()

fig.savefig("avg_goals.png")
```
